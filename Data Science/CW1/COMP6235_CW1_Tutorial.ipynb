# Digging deeper into the correlation

```python
# Create a regplot to figure out the correlation

sns.regplot(x="Time", y="Weight", data=time_weight_bait_data)
```

```python
#Import stats from scipy

from scipy import stats
```

```python
# Use the to find the correlation details 
# Is it positive or negative?
# Is it statistically significant?

stats.pearsonr(time_weight_bait_data["Time"], time_weight_bait_data["Weight"])
# returns correlation coefficient and p-value
# The P-value is the probability that you would have found the current result 
# if the correlation coefficient were in fact zero (null hypothesis)

# Assume that P-value < 0.05, is a statistically significant
# but the correlation is a weak correlation
```

- Negative Correlation with correlation coefficient = -0.1205929
- P-value = 0.0158, confidence level ~= 98%

```python
# Define a named function that calculates the R-squared using what we did in the previous step

def r2(x,y):
    return stats.pearsonr(x, y)[0]**2
```

```python
r2(time_weight_bait_data["Time"], time_weight_bait_data["Weight"])
```

```python
# You can experiment with some other functions like jointplot

sns.jointplot(x="Time", y="Weight", data=time_weight_bait_data, kind="reg", stat_func=r2)
```
